В данном домашнем задании предлагалось провести сборку генома с другими параметрами. Главным параметром, который варьировался на семинаре, была длина контига. Были опробованы сборки с контигами длиной 39, 49 и 65 пар нуклеотидов. В качестве референса была выбрана длина в 49 пар. Проверялась сборка на близком значении - 45, и на двух сильно отклонённых значениях, в меньшую - 25, и в большую - 85 пар нуклеотидов. Критерием качества сборки считалась похожесть выходных данных на скаффолд, полученный на семинаре.
